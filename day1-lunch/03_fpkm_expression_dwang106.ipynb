# Question 1: Plot sex as series

1. Plot a time course of FBtr0331261 with each sex is a separate series.

#### Load the goi and separate the series by sex

```python
goi = "FBtr0331261"
df_sxl = df_fpkms.loc[goi,:]

# separate the seires by sex by partitioning the series
m = df_sxl[1:9]
f = df_sxl[10:]

print(m)
print(f)
```

#### Plot the time series by sex

```python
fig, ax = plt.subplots(2)

fig.set_figheight(10)
ax[0].plot(m.index,m)
ax[0].set_title('FBtr0331261 Male Time Series')
ax[0].set_xlabel('Gene Name')
ax[0].set_ylabel('Sxl')

ax[1].set_title('FBtr0331261 Female Time Series')
ax[1].plot(f.index,f)
ax[1].set_xlabel('Gene Name')
ax[1].set_ylabel('Sxl')
```
