# Question 1: Plot sex as series

1. Plot a time course of FBtr0331261 with each sex is a separate series.

#### Load the goi and separate the series by sex

```python
goi = "FBtr0331261"
df_sxl = df_fpkms.loc[goi,:]

# separate the series by sex by partitioning the series
m = df_sxl[1:9]
f = df_sxl[10:]

print(m)
print(f)
```

#### Plot the time series by sex

```python
timeFrame = ['10','11','12','13','14A','14B','14C','14D']

plt.plot(timeFrame,m,label='Male')
plt.plot(timeFrame,f,label='Female')
plt.legend(loc='upper right',fontsize=20)
plt.rcParams['xtick.labelsize']=20
plt.rcParams['ytick.labelsize']=20
plt.xlabel('Time',fontsize=20)
plt.ylabel('Gene Name',fontsize=20)
title = "FBtr0331261 Time Series"
plt.title(title,fontsize=24)
plt.show()
```
