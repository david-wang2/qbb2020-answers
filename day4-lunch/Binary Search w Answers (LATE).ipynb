```python
# this recursive binary search is clearly my own work because
# it is much less elegant than the binary search produced elsewhere
# this binary search is lowkey bad..... because it uses one more index position than a normal recursive binary search

def recursive_binary(search, df, low, high, numIter, prevIndex):
    # base cases:
    if (high - low < 2):
        return (prevIndex,numIter)
 
    if (df.iloc[prevIndex]["Start"] < search and df.iloc[prevIndex]["End"] > search):
        return (prevIndex,numIter)
    
    numIter += 1
    index = low + int(abs(high-low)/2)
    # recursive case
    if (df.iloc[index]["Start"] < search):
        return recursive_binary(search,df,index,high,numIter,index)
    elif (df.iloc[index]["End"] > search):
        return recursive_binary(search,df,low,index,numIter,index)
        
i, count = recursive_binary(21378950, df, 0, df.shape[0]-1, 0, 0)
```
