```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from scipy import stats
import statsmodels.formula.api as smf
import statsmodels.api as sm
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2020/2020-03-10/salary_potential.csv')
```

```python
roi = df["state_name"] == "North-Carolina"
df_nc = df.loc[roi, :]
```

```python
df_nc
```

```python
model = smf.ols(formula = "make_world_better_percent ~ stem_percent ", data = df)
results = model.fit()
results.summary()
```

```python
plt.scatter(df["stem_percent"], df["make_world_better_percent"])
plt.title("Stem Percent vs Make World Better Place")
plt.xlabel("Stem Percent")
plt.ylabel("Make World Better Place")
plt.show()
```
